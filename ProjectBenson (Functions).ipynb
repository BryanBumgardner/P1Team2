```python
import collections
import csv

%matplotlib inline
import matplotlib.pyplot as plt

from collections import defaultdict
from datetime import datetime
```

```python
def readdata(filename):

    tempdict = defaultdict(list)
    
    with open(filename,) as csvfile:
        temp = csv.reader(csvfile, delimiter=',')
        next(temp, None)
        
        for row in temp:
            row[10] = row[10].rstrip()
            tempdict[tuple(row[:4])].append(tuple(row[4:]))

    return tempdict
```

```python
def tidydata(temp):
    
    tempdict = defaultdict(list)
    
    tempdict = {k : [(datetime.strptime(date + time,'%m/%d/%Y%X'), int(cumin))
                                       for _, _, date, time,_, cumin, _ in val]
                                       for k, val in temp.items()}

    for val in tempdict.values():
        val = sorted(val)

    return tempdict
```

```python
def turns(temp):
    
    counts = defaultdict(list)
    dailytotal = defaultdict(int)
    tempdict = defaultdict(int)

    counts = {k : [[v[i][0], v[i+1][1] - v[i][1], v[i+1][0] - v[i][0]]
                   for i in range(len(v)-1)]
                   for k, v in temp.items()}
    
    for k, val in counts.items():
        
        for time, count, delta in val:
            day = time.date()
            tempdict[day] += count
        
        dailytotal[k] = sorted(tempdict.items())

    return dailytotal
```

```python
def plots(temp):
    
    dates = []
    counts = [] 
    
    for k,val in temp.items():
        
        dates = []
        counts = []
        
        for i in range(0,len(val)):
            dates.append(v[i][0])
            counts.append(v[i][1])
    
        ##  Plotting the data
        plt.figure(figsize=(10,3))
        plt.plot(dates,counts)
        plt.xlabel('Date')
        plt.ylabel('Daily Turnstile Counts')
        plt.title('New York City MTA Subway Turnstile Counts')
    
    plt.show()
```

```python
def station(temp):
    
    stationtotal = defaultdict(int)
    
    for k, val in temp.items():
        for i in range(0,len(val)):
            stationtotal[k[3]] += val[i][1] 
        
    return stationtotal
```

## Main

```python
rawdata = defaultdict(list)
entrancedata = defaultdict(list)
#filename = 'turnstile_150523.txt'
#filename = 'turnstile_150530.txt'
filename = 'turnstile_150606.txt'
#filename = 'turnstile_150613.txt'

rawdata = readdata(filename)

entrancedata = tidydata(rawdata)

entrancedata = turns(entrancedata)

#plots(entrancedata)

stationentrancedata = station(entrancedata)

#print entrancedata
print stationentrancedata
```
